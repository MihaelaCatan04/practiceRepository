##**Violin Plots**


Why use violin plots? They place more emphasis on your data's distrbution.

###Create a violin plot from the DataFrame showing the Kernel density estimation of the tips data.
Notice the bars inside each "violin" are similar to box plots.

```python
sns.violinplot(data=categorical_data, x='day', y='tip')
```

###Create a violin plot like above, but make the broadness of each "violin" proportional to the number of counts.

```python
sns.violinplot(data=categorical_data, x='day', y='tip',scale='count')
```

###Create a violin plot like above, but change how quartiles are represented.
Notice how the whiskers have been removed, leaving only the 1st, 2nd, and 3rd quartiles.

```python
sns.violinplot(data=categorical_data, x='day', y='tip',inner='quartile')
```

###Create a violin plot like above, but display points instead of summary statistics.

```python
sns.violinplot(data=categorical_data, x='day', y='tip',inner='point')
```

###Create a double-violin plot where tips are seperated by sex.

```python
sns.violinplot(data=categorical_data, x='day', y='tip', hue='sex')
```

###Create a double-violin plot like above but where "violin-halves" are merged.
Notice how the whiskers have been lost.

```python
sns.violinplot(data=categorical_data, x='day', y='tip',hue='sex',split='True')
```

##**Selecting Plot Styles**

###Change the aesthetics of the combined plot.

```python
sns.set_style("white")
sns.set_context("poster")
axes=sns.boxplot(data=categorical_data, x='day', y='tip',fliersize=0,boxprops=dict({'edgecolor': 'black', 'facecolor': (1,1,1,0)}))
axes=sns.stripplot(data=categorical_data, x='day', y='tip',alpha=.5)
axes.set(xlabel='Day', ylabel='Tip')
sns.despine()
```

###Change the aesthetics of the violin plot.

```python
sns.set_style("white")
sns.set_context("poster")
axes2=sns.violinplot(data=categorical_data, x='day', y='tip')
axes2.set(xlabel='Day', ylabel='Tip')
sns.despine()
```

Notice how both these figures are now easier to read from afar. This is in large part due to *sns.set_context("poster")* .

Other contexts include 'notebook' (default), 'paper', and 'talk'.

#**Customizing Beyond Seaborn's Defaults**

##**Ordering Plot Elements**

###Create a bar plot showing the mean tip on each day in a custom order.
Notice how a single line of code could become very long with 10+ elements.

```python
sns.barplot(data=categorical_data, x='day', y='tip', order=['Fri', 'Thur', 'Sat', 'Sun'])
```

###Create the same bar plot as above, but by passing the order list by name.

```python
x_order = ['Fri', 'Thur', 'Sat', 'Sun']
sns.barplot(data=categorical_data, x='day', y='tip', order=x_order)
```

###Create a double-bar plot like above, but where the sub-category of *day* follows a custom order.

```python
sns.barplot(data=categorical_data, x='sex', y='tip', hue_order=x_order,hue='day')
```

##**Customizing Color Palettes**

###Create a bar plot like above where a Python dictionary designates custom colors.
The full range of supported colors can be viewed here: [ List of named colors ](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
color_dict={'Thur':'purple','Fri':'orange','Sat':'brown','Sun':'gray'}
sns.barplot(data=categorical_data, x='day', y='tip',palette=color_dict)
```

#**Plotting matrix data with Seaborn**

##**Reshaping DataFrames**

###Load in a DataFrame containing categorical data. This DataFrame has columns for the year and month different flights and the number of passengers on each flight.

```python
categorical_data = sns.load_dataset('flights')
categorical_data.head()
```

###Convert the categorical data to matrix data.

```python
matrix_data = categorical_data.pivot(index='month', columns='year', values='passengers')
matrix_data.head()
```

```python
matrix_data = categorical_data.pivot(index='month', columns='year', values='passengers')
matrix_data.head()
```

##**Heatmaps**

###Create a heatmap of the matrix data displaying the number of passengers across the months of several years.

```python
sns.heatmap(data=matrix_data)
```

###Create a heatmap like above but annotate each colored square with its respective numerical value.

```python
sns.heatmap(data=matrix_data, annot=True)
```

###Create a heatmap like above but annotate with decimal numerals instead of floating point numbers.

```python
sns.heatmap(data=matrix_data, annot=True,fmt='d')
```

###Create a heatmap like above but change the color scheme.

```python
sns.heatmap(data=matrix_data, annot=True,fmt='d',cmap='coolwarm')
```

###Plot #5

While the scatter plot gave us a general idea what passengers of different ages and boarding classes paid, it was unable to provide us much more demographic-specific information beyond that.

While we could make a box+strip plot or violin plot for each combination of boy/girl/man/woman and first/second/third classes, the resulting graph would probably look very busy.

So instead, let's make use a heatplot. We can start by running the following code:

As you copy-paste it, try and follow what the commands do based on what you learned during the first lecture on DataFrame manipulation with Pandas.

```python
lab3_data_subset=lab3_data[['class','who','fare']]
lab3_data_subset=lab3_data_subset.dropna(how='any')
lab3_data_subsetAverage=lab3_data_subset.groupby(['class','who']).mean()
lab3_data_subsetAverage=lab3_data_subsetAverage.reset_index()
lab3_data_subsetAverage.head()
```

Next, convert this categorical data into matrix data as was shown in the tutorial. The new DataFrame containing the matrix data should be named 'matrix_data'. Once completed, run the following code to plot the heatmap.

```python
matrix_data = lab3_data_subsetAverage.pivot(index='class', columns='who', values='fare')
matrix_data
```

```python
axes5 = sns.heatmap(data=matrix_data, annot=True,fmt='.2f',cmap='coolwarm')
for text in axes5.texts: text.set_text('£' + text.get_text())
axes5.set(xlabel='Demographic', ylabel='Boarding Class', xticklabels=['boy','girl','man','woman'])
plt.savefig("/content/drive/My Drive/Figure4.png", bbox_inches='tight')
```

image.png

As we did with the scatter plot, we can see that Second Class and Third Class passengers generally paid around the same amount while individuals in First Class had relatively higher variation in their fares. However, with the heatmap we can now clearly see that within the First Class, girls paid the highest average fare while First Class men paid the least.

Furthermore, we can obtain numerical values for each category directly from the plot itself. This lends itself well toward converting fares from pound sterling in 1912 to their modern-day value.
